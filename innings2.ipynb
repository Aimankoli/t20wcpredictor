```python
balls_innings2 = pd.DataFrame()
#dataframe for team in first innings, we will make one for second innings too
ctr = 1
dfs = []
for index, row in tqdm(m.iterrows()):
    if ctr in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        ctr +=1
        continue
    ctr+=1
    ball_no = []
    batsman = []
    bowler = []
    runs = []
    player_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    wickets = []
    try:
        for ball in row['innings'][1]['2nd innings']['deliveries']:
            for key in ball.keys():
                match_id.append(ctr)
                batting_team.append(row['innings'][1]['2nd innings']['team'])
                teams.append(row['info.teams'])
                ball_no.append(key)
                batsman.append(ball[key]['batsman'])
                bowler.append(ball[key]['bowler'])
                runs.append(ball[key]['runs']['total'])
                city.append(row['info.city'])
                venue.append(row['info.venue'])
                try:
                    player_dismissed.append(ball[key]['wicket']['player_out'])
                except:
                    player_dismissed.append('0')
                try: 
                    if (ball[key]['wicket']):
                        wickets.append(1)
                except:
                    wickets.append(0)
    
            iterdf = pd.DataFrame({
                'match_id':match_id,
                'teams':teams,
                'batting_team':batting_team,
                'ball':ball_no,
                'batsman':batsman,
                'bowler':bowler,
                'runs':runs,
                'player_dismissed':player_dismissed,
                'city':city,
                'venue':venue,
                'wicket': wickets
            })
            
        dfs.append(iterdf)
    
        balls_innings2 = pd.concat(dfs, ignore_index=True)
    except:
        continue
```
