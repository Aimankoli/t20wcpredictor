```python
import pandas as pd
import pickle
import numpy as np
```

```python
b2 = pickle.load(open('rawi2.pkl', 'rb'))
b2
```

```python
def bowlteam(row):
    for team in row['teams']:
        if team!=row['batting_team']:
            return team
b2['bowling_team'] = b2.apply(bowlteam, axis = 1)
```

```python
b2.drop(columns = ['teams'], inplace = True)
```

```python
teams = {
    'Pakistan',
    'South Africa',
    'India',
    'New Zealand',
    'Sri Lanka',
    'West Indies',
    'Australia',
    'England',
    'Afghanistan',
    'Bangladesh',
    'Ireland',
    'Zimbabwe',
    'Netherlands',
    'United Arab Emirates',
    'Scotland',
    'Oman',
    'Nepal',
    'Papua New Guinea',
    'Canada',
    'Namibia',
    'United States of America',
    'Uganda'
    
    
}
```

```python
b2=b2[b2['batting_team'].isin(teams)]
b2=b2[b2['bowling_team'].isin(teams)]
```

```python
fb2 = b2[['match_id', 'batting_team', 'bowling_team', 'ball', 'runs', 'player_dismissed', 'wicket', 'city', 'venue']]
```

```python
cities = np.where(fb2['city'].isnull(), fb2['venue'].str.split().apply(lambda x:x[0]), fb2['city'])
```

```python
fb2['city'] = cities
```

```python
fb2
```

```python
fb2.drop(columns = ['venue'], inplace = True)
```

```python
eligible_cities = fb2['city'].value_counts()[fb2['city'].value_counts() > 600].index.tolist()

fb2 = fb2[fb2['city'].isin(eligible_cities)]
```

```python
fb2
```

```python
# Convert 'runs' column to numeric type
fb2['runs'] = pd.to_numeric(fb2['runs'], errors='coerce')

# Group by 'match_id' and calculate cumulative sum for 'runs'
fb2['cumulative_runs'] = fb2.groupby('match_id')['runs'].cumsum()
```

```python
# Convert 'runs' column to numeric type
fb2['wicket'] = pd.to_numeric(fb2['wicket'], errors='coerce')

# Group by 'match_id' and calculate cumulative sum for 'runs'
fb2['cumulative_wickets'] = fb2.groupby('match_id')['wicket'].cumsum()
```

```python
fb2
```

```python
fb2['over'] = fb2['ball'].apply(lambda x:str(x).split(".")[0])
fb2['ballnum'] = fb2['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
fb2['wicketsleft'] = fb2['cumulative_wickets'].apply(lambda x: 10-x)
```

```python
fb2['ballsleft'] = 120 - (fb2['over'].astype('int')*6 + fb2['ballnum'].astype('int'))
```

```python
fb2['ballsleft'] = fb2['ballsleft'].apply(lambda x:0 if x<0 else x)
```

```python
fb2
```

```python
fb2['rr'] = fb2['cumulative_runs'] / ((120-fb2['ballsleft'])/6)
```

```python
fb2
```

```python
# Convert the 'runs' column to numeric, coercing errors to NaN
fb2['runs'] = pd.to_numeric(fb2['runs'], errors='coerce')

# Group by 'match_id'
matchgroups = fb2.groupby('match_id')
match_ids = fb2['match_id'].unique()
last5 = []

# Iterate over each match_id
for id in match_ids:
    group = matchgroups.get_group(id)
    
    # Perform the rolling sum operation
    rolling_sum = group['runs'].rolling(window=30).sum()
    
    # Extend the list with the rolling sum values
    last5.extend(rolling_sum.values.tolist())
last5w = []
fb2['wicket'] = pd.to_numeric(fb2['wicket'], errors='coerce')
matchgroups = fb2.groupby('match_id')
match_ids = fb2['match_id'].unique()
for id in match_ids:
    group = matchgroups.get_group(id)
    
    # Perform the rolling sum operation
    rolling_sum = group['wicket'].rolling(window=30).sum()
    
    # Extend the list with the rolling sum values
    last5w.extend(rolling_sum.values.tolist())
```

```python
fb2['last5'] = last5
fb2['last5w'] = last5w
```
