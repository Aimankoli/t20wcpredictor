```python
import pandas as pd
import pickle
import numpy as np
```

```python
df = pickle.load(open('rawi2-2.pkl', 'rb'))
t = pickle.load(open('id&runs','rb'))
```

```python
df
```

```python
t
```

```python
df = df.merge(t, on='match_id', how='left')
```

```python
df
```

```python
def rrr(row):
    for runs in ['runs_x']:
        rr = row['runs_x'] - row['cumulative_runs']
        oversleft = row['ballsleft']/6
        if oversleft == 0:
            if row['cumulative_runs']>=row['runs_x']:
                return 0
            else:
                return -1
        return rr/oversleft
    
```

```python
df['rrr'] = df.apply(rrr, axis = 1)
```

```python
df
```

```python
df['rrr'] = df['rrr'].apply(lambda x: 0 if x<0 else x)
```

```python
df
```

```python
df.rename(columns={'runs_x': 'target'}, inplace=True)
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df, on = 'match_id')
final_df
```

```python
last_rr = df.groupby('match_id')['rr'].last().reset_index()
```

```python
last_rr
```

```python
last_rr.rename(columns={'rr': 'final_rr'}, inplace=True)
final_df = final_df.merge(last_rr, on='match_id', how='left')
```
