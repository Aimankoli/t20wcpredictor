```python
import pickle
m = pickle.load(open('dataset.pkl', 'rb'))
m
```

```python
import pandas as pd
from yaml import safe_load
import numpy as np
import os
from tqdm import tqdm
```

```python
balls_innings1 = pd.DataFrame()
#dataframe for team in first innings, we will make one for second innings too
ctr = 1
dfs = []
for index, row in m.iterrows():
    if ctr in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        ctr +=1
        continue
    ctr+=1
    ball_no = []
    batsman = []
    bowler = []
    runs = []
    player_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    wickets = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(ctr)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_no.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_dismissed.append('0')
            try: 
                if (ball[key]['wicket']):
                    wickets.append(0)
            except:
                wickets.append(0)
    
    iterdf = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_no,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_dismissed,
            'city':city,
            'venue':venue,
            'wicket': wickets
        })
    dfs.append(iterdf)
    balls_innings1 = pd.concat(dfs, ignore_index=True)
    
```

```python
balls_innings1['wicket'] = 1
```

```python
balls_innings1['wicket'].value_counts()
```

```python
for index, row in balls_innings1.iterrows():
    # Check if the value in player_dismissed is '0'
    if row['player_dismissed'] == '0':
        balls_innings1.at[index, 'wicket'] = 0
```

```python
#wickets data is ready
pickle.dump(balls_innings1, open('b1.pkl', 'wb'))
```
