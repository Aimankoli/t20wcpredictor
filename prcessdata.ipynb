```python
cities = np.where(df['city'].isnull(), df['venue'].str.split().apply(lambda x:x[0]), df['city'])
```

```python
df.drop(columns = ['venue'], inplace = True)
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()

df = df[df['city'].isin(eligible_cities)]
```

```python
df.dtypes
```

```python
# Convert 'runs' column to numeric type
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')

# Group by 'match_id' and calculate cumulative sum for 'runs'
df['cumulative_runs'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
# Convert 'runs' column to numeric type
df['wicket'] = pd.to_numeric(df['wicket'], errors='coerce')

# Group by 'match_id' and calculate cumulative sum for 'runs'
df['cumulative_wickets'] = df.groupby('match_id')['wicket'].cumsum()
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ballnum'] = df['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
df
```

```python
df['wicketsleft'] = df['wicket'].apply(lambda x: 10-x)
```

```python
df
```

```python
df['ballsleft'] = 120 - (df['over'].astype('int')*6 + df['ballnum'].astype('int'))
```

```python
df
```

```python
df['ballsleft'] = df['ballsleft'].apply(lambda x:0 if x<0 else x)
```

```python
df
```

```python
df['rr'] = df['cumulative_runs'] / ((120-df['ballsleft'])/6)
```

```python
df
```

```python
# Convert the 'runs' column to numeric, coercing errors to NaN
df['runs'] = pd.to_numeric(df['runs'], errors='coerce')

# Group by 'match_id'
matchgroups = df.groupby('match_id')
match_ids = df['match_id'].unique()
last5 = []

# Iterate over each match_id
for id in match_ids:
    group = matchgroups.get_group(id)
    
    # Perform the rolling sum operation
    rolling_sum = group['runs'].rolling(window=30).sum()
    
    # Extend the list with the rolling sum values
    last5.extend(rolling_sum.values.tolist())
```

```python
last5w = []
df['wicket'] = pd.to_numeric(df['wicket'], errors='coerce')
matchgroups = df.groupby('match_id')
match_ids = df['match_id'].unique()
for id in match_ids:
    group = matchgroups.get_group(id)
    
    # Perform the rolling sum operation
    rolling_sum = group['wicket'].rolling(window=30).sum()
    
    # Extend the list with the rolling sum values
    last5w.extend(rolling_sum.values.tolist())
```

```python
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df, on = 'match_id')
final_df
```

```python
final_df['last5'] = last5
final_df['last5w'] = last5w
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
#random forest not giving accurate results, try xgb
```
